```python
def is_possible(game_info, red_count, green_count, blue_count):
    for subset in game_info:
        print(subset)
        string_subset = str(subset)
        cubes_subset = string_subset[1:-1].split('; ')
        for cube_set in cubes_subset:
            cubes = cube_set.split(', ')
            
            for cube in cubes:
                count, color = cube.split()
                
                count = int(count.strip("'"))
                if color == 'red' and count > red_count:
                    return False
                elif color == 'green' and count > green_count:
                    return False
                elif color == 'blue' and count > blue_count:
                    return False
                else:
                    red_count -= count if color == 'red' else 0
                    green_count -= count if color == 'green' else 0
                    blue_count -= count if color == 'blue' else 0
    return True

def possible_games(game_records, red_count, green_count, blue_count):
    possible_game_ids = []
    for game_record in game_records:
        game_id, cubes_info = game_record.split(': ')
        cubes_sets = cubes_info.split('; ')
        game_info = [subset.split(', ') for subset in cubes_sets]
        print(game_info)
        if is_possible(game_info, red_count, green_count, blue_count):
            possible_game_ids.append(int(game_id.split()[-1]))

    return possible_game_ids

# Example game records
games = [
    "Game 1: 3 blue, 4 red; 1 red, 2 green, 6 blue; 2 green",
    "Game 2: 1 blue, 2 green; 3 green, 4 blue, 1 red; 1 green, 1 blue",
    "Game 3: 8 green, 6 blue, 20 red; 5 blue, 4 red, 13 green; 5 green, 1 red",
    "Game 4: 1 green, 3 red, 6 blue; 3 green, 6 red; 3 green, 15 blue, 14 red",
    "Game 5: 6 red, 1 blue, 3 green; 2 blue, 1 red, 2 green"
]

# Cube counts
red_count = 12
green_count = 13
blue_count = 14

possible_ids = possible_games(games, red_count, green_count, blue_count)
sum_of_possible_ids = sum(possible_ids)

print(f"The IDs of the possible games are: {possible_ids}")
print(f"The sum of the IDs of the possible games is: {sum_of_possible_ids}")
```
